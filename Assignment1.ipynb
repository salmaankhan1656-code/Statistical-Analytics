```python
age_groups = [(15, 25), (25, 35), (35, 45), (45, 55), (55, 65)]
frequency = [350, 375, 120, 80, 90]
class_marks = [(a + b) / 2 for a, b in age_groups]
total_f = sum(frequency)
sum_fx = sum(f * x for f, x in zip(frequency, class_marks))
mean = sum_fx / total_f
cf = []
cumulative = 0
for f in frequency:
    cumulative += f
    cf.append(cumulative)
n = total_f
median_class_index = next(i for i, c in enumerate(cf) if c >= n/2)
L = age_groups[median_class_index][0]
f = frequency[median_class_index]
cf_prev = 0 if median_class_index == 0 else cf[median_class_index - 1]
h = age_groups[median_class_index][1] - age_groups[median_class_index][0]
median = L + ((n/2 - cf_prev) / f) * h
modal_class_index = frequency.index(max(frequency))
L = age_groups[modal_class_index][0]
f1 = frequency[modal_class_index]
f0 = 0 if modal_class_index == 0 else frequency[modal_class_index - 1]
f2 = 0 if modal_class_index == len(frequency) - 1 else frequency[modal_class_index + 1]
h = age_groups[modal_class_index][1] - age_groups[modal_class_index][0]
mode = L + ((f1 - f0) / ((2 * f1) - f0 - f2)) * h
print(f"Mean = {mean:.2f}")
print(f"Median = {median:.2f}")
print(f"Mode = {mode:.2f}")
```
